```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder, StandardScaler
import seaborn as sns

data = pd.read_csv('data/insurance.csv')
data.head()
data.shape
data.info()
data.describe()
```

1. Mengidentifikasi variabel-variabel yang akan digunakan sebagai variabel bebas (fitur) dan variabel target (biaya medis personal)
- Variabel Bebas  : Age, Sex, Bmi, childern, smoker, region
- Variabel Target : charges 

2. Membagi dataset menjadi data latih dan data uji (test) dengan proporsi yang sesuai

```python
# Membuat variabel bebas X dan Y, contoh pengambilan dari analisis korelasi sebelumnya
X = data[['age', 'bmi', 'children']] # Kolom fitur
y = data['charges'] # Kolom target

# Pembagian data latih dan data uji dengan proporsi 7:3
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.7, test_size=0.3, random_state=100)

print("Jumlah data latih: ", len(X_train))
print("Jumlah data uji: ", len(X_test))
```

```python
# Visualisasi data dengan pairplot
sns.pairplot(data, x_vars=['age', 'bmi','children'],
             y_vars='charges', size=4, aspect=1, kind='scatter')
plt.show()
```

```python
# Visualisasi korelasi dengan heatmap
sns.heatmap(data.select_dtypes(include=np.number).corr(), cmap="YlGnBu", annot = True)
plt.show()
```

3. Melakukan feature scaling

```python
# Feature scaling.
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

4. Membuat model multiple linear regression menggunakan Scikit-Learn.

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train, y_train)
```

5. Latih model pada data latih dan melakukan prediksi pada data uji.

```python
#Training Model
import statsmodels.api as sm

X_train_sm = sm.add_constant(X_train)
lr = sm.OLS(y_train, X_train_sm).fit()

# Prediksi pada data uji
X_test_sm = sm.add_constant(X_test)
Y_test_pred = lr.predict(X_test_sm)

plt.scatter(y_test, Y_test_pred)
plt.xlabel("Nialai Aktual")
plt.ylabel("Nilai Prediksi")
plt.title("Scatter Plot: Nilai Aktual vs. Nilai Prediksi")
plt.show()
```

6. Evaluasi model dengan menghitung metrik seperti R-squared, MSE, dan MAE. Tampilkan hasil evaluasi.
Penejelasan : 
Pada kode program dibawah ini menggunakan model linear regression menggunakan matriks evaluasi.

```python
# Evaluasi model
from sklearn.metrics import mean_absolute_error, mean_squared_error

# Hitung nilai R-squad
from sklearn.metrics import r2_score

mae = mean_absolute_error(y_test, Y_test_pred)

# Menghitung MSE
mse = mean_squared_error(y_test, Y_test_pred)

# Menghitung RMSE
rmse = np.sqrt(mse)

# Menghitung R-squared
r2 = r2_score(y_test, Y_test_pred)

print("MAE:", mae)
print("MSE:", mse)
print("RMSE:", rmse)
print("R-squared:", r2)
```

```python
# Tampilkan visualisasi data yang relevan seperti scatter plot atau grafik lainnya jika diperlukan untuk pemahaman yang lebih baik.
plt.scatter(y_test, Y_test_pred)
plt.plot(y_test, y_test, color='red')
plt.xlabel('Actual Charges')
plt.ylabel('Predicted Charges')
plt.title('Actual Charges vs Predicted Charges')
plt.show()
```
